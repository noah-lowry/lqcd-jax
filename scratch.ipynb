```python
import re
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

values = np.array([float(re.findall("o=(-?\d*\.?\d*)", s)[0]) for s in open("data.txt").read().splitlines()])
```

```python
pt_func = lambda t: np.cov(values[:-t], values[t:])[0, 1] / np.cov(values, values)[0, 1]
acorr_series = np.array([pt_func(i) for i in 1+np.arange(1000)])
```

```python
plt.plot(1+2*np.cumsum(acorr_series))
```
